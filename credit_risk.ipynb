```python
cr.Gender.fillna('Male', inplace=True)
cr.Married.fillna('Yes', inplace=True)
cr.Dependents.fillna('0', inplace=True)
cr.Occupation.fillna('No', inplace=True)
cr.LoanAmount.fillna(cr.LoanAmount.mean(), inplace=True)
cr.Loan_Amount_Term.fillna(cr.Loan_Amount_Term.mean(), inplace=True)
cr.Credit_History.fillna(0, inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
cr = cr.drop(['Loan_ID'], axis=1)
```

```python
# This line replaces all missing values (NaN) in text (categorical) columns with the word "Missing"
cr[cr.select_dtypes(include='object').columns] = (
    cr.select_dtypes(include='object')
      .fillna('Missing')
)
```

```python
# This line converts all text (object) columns into pure string format.
cr[cr.select_dtypes(include='object').columns] = (
    cr.select_dtypes(include='object')
      .astype(str)
)
```

```python
cr[cr.select_dtypes(include='object').columns] = cr[cr.select_dtypes(include='object').columns].apply(le.fit_transform)
```

```python
cr
```

```python
from sklearn.model_selection import train_test_split
cr_train, cr_test = train_test_split(cr, test_size=0.2, random_state=42)
cr_train_x = cr_train.iloc[:,0:-1]
cr_train_y = cr_train.iloc[:,-1]
cr_test_x = cr_test.iloc[:,0:-1]
cr_test_y = cr_test.iloc[:,-1]
```

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()
log_reg.fit(cr_train_x, cr_train_y)
```

```python
pred = log_reg.predict(cr_test_x)
```

```python
act_pred = pd.DataFrame({'Actual': cr_test_y, 'Predicted': log_reg.predict(cr_test_x)})
act_pred
```
